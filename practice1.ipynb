```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel(r'H:\Education\4th year 2nd semester\PME 00\excel\Book1.xlsx')
```

```python
df.columns=(['Corrected depth', 'total nmr porosity', 'capillary water BVI', 'NMR K', 'core K'])
```

```python
df.columns
```

```python
sns.set_style('darkgrid')
fig0 =sns.pairplot(df,  x_vars=['Corrected depth', 'total nmr porosity', 'capillary water BVI', 'NMR K'], 
                  y_vars=['core K'])
fig0.fig.set_size_inches(15,5)
fig0.savefig("output.png")
```
